```python
# ======================================
# EDA_Final_Project.ipynb
# Exploratory Data Analysis (EDA) using built-in dataset
# ======================================

# ===============================
# 1. Import Libraries
# ===============================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# ===============================
# 2. Load Built-in Dataset
# ===============================
# Using Seaborn 'tips' dataset
data = sns.load_dataset('tips')
data.head()
```

```python
# ===============================
# 3. Data Cleaning
# ===============================
# Check for missing values
print(data.isnull().sum())

# Info about dataset
data.info()
```

```python
# ===============================
# 4. Exploratory Data Analysis
# ===============================

# Summary statistics
data.describe()
```

```python
# Distribution of total bill
plt.figure(figsize=(6,4))
sns.histplot(data['total_bill'], bins=20, kde=True, color='skyblue')
plt.title('Total Bill Distribution')
plt.xlabel('Total Bill')
plt.show()
```

```python
# Tip vs Total Bill scatter plot
plt.figure(figsize=(6,4))
sns.scatterplot(x='total_bill', y='tip', hue='day', data=data, s=100)
plt.title('Tip vs Total Bill by Day')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# Select only numeric columns
numeric_data = data.select_dtypes(include='number')

# Plot correlation heatmap
plt.figure(figsize=(6,4))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# ===============================
# 5. Key Findings
# ===============================

- Total bill ranges from \$3\$50, most bills are under \$20.
- Tips increase with total bill, as expected.
- Saturday and Sunday have higher total bills on average.
- Theres a positive correlation between total_bill and tip (~0.67).
